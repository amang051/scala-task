```scala
case class Student(sno: Int, name: String, score: Int)

implicit class Conversion(val list : List[Student]) {
    def %>(score : Int) : List[Student] = list.filter(student => student.score > score)
    def %<(score : Int) : List[Student] = list.filter(student => student.score < score)
}

val sList = List(
    Student(1, "Aman", 80),
    Student(2, "Ajay", 50),
    Student(3, "Rahul", 60),
    Student(4, "Vijay", 55),
    Student(5, "Rakesh", 75),
    Student(6, "Dinesh", 45)
)

val list1 = sList %> 70
val list2 = sList %< 60

println(list1)
println(list2)
```

```scala
import scala.language.implicitConversions
import scala.collection.mutable.ListBuffer
import scala.collection.immutable.ArraySeq

implicit def tupleToStudent(t: (Int, String, Int)) : Student = Student(t._1, t._2, t._3)

class StudentOps(stuList : ListBuffer[Student]) {
    def printStudent : Unit = {
        println(stuList)
    }

    def add(student : Student) : Unit = {
        stuList += student
    }

    def filterStudents(x : Student => Boolean) : ListBuffer[Student] = {
        stuList.filter(x)
    }
}

val sList = ListBuffer(
    Student(1, "Aman", 80),
    Student(2, "Ajay", 50),
    Student(3, "Rahul", 60),
    Student(4, "Vijay", 55),
    Student(5, "Rakesh", 75),
    Student(6, "Dinesh", 45)
)

val studentObj: StudentOps = new StudentOps(sList)
studentObj.printStudent
studentObj.add(new Student(7, "Suresh", 78))
studentObj.printStudent
studentObj.add((8, "Sumit", 54))
studentObj.printStudent
```

```scala
implicit class Filter(studentArr : Array[Student]) {
    def filterRecords(x : Student => Boolean) : Array[Student] = {
        studentArr.filter(x)
    }
}

val sList: Array[Student] = Array(
    Student(1, "Aman", 80),
    Student(2, "Ajay", 50),
    Student(3, "Rahul", 60),
    Student(4, "Vijay", 55),
    Student(5, "Rakesh", 75),
    Student(6, "Dinesh", 45)
)

val res = sList.filterRecords(student => student.score > 50)
res.foreach(println)
```
